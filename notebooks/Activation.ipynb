Plot the activation function new design 

```python
import plotly.graph_objects as go

act = 'reflectexpint'

def plot(x, act, title):
    fig = go.Figure()
    f = act_dict[act].forward(x)
    f1 = grad_dict[act_dict[act].__class__.__name__][0](f, x)
    f2 = grad_dict[act_dict[act].__class__.__name__][1](f, x)
    f3 = grad_dict[act_dict[act].__class__.__name__][2](f, x)
    fig.add_trace(go.Scatter(x=x, y=f,  mode='lines',  name='activation'))
    fig.add_trace(go.Scatter(x=x, y=f1, mode='lines', name='1st derivative'))
    fig.add_trace(go.Scatter(x=x, y=f2, mode='lines', name='2nd derivative'))
    fig.add_trace(go.Scatter(x=x, y=f3, mode='lines', name='3rd derivative'))

    mathjax_link = 'https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js'
    fig.update_layout(
        title=title,
        xaxis_title="x",
        yaxis_title="f"
    )
    fig.show()

x = torch.arange(-3., 3., 0.001)
plot(x, 'reflectexp', 'Reflection Exponential Activation')
```
